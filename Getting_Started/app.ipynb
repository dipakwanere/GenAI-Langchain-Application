```python
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
GROQ_API_KEY = os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(model="llama-3.1-8b-instant", groq_api_key = GROQ_API_KEY)
```

```python
from langchain_core.messages import HumanMessage,SystemMessage
messages = [SystemMessage(content="Translate the following message from English to Spanish and Hindi."),
            HumanMessage(content="Hellow, How are you?")]

response = model.invoke(messages)
print(response.content)
```

```python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
parser.invoke(response)    # this will return the content as a string
```

```python
# now let's chain the modela nd parser with Pipe character |

chain = model | parser
chain.invoke(messages)  # returns the content as a string
```
